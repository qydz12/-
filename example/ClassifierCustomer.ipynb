### 分析客户人群分布以及平均消费金额

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#读取数据
df = pd.read_csv('D:/study/Python/data_analyst/data_clean/zxc.csv')
```

### 客群分布柱状图

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=df, x="customer_count", y="Segment", palette="Blues_d",hue="Segment", legend=False)
plt.title("RFM 客户分群分布",fontsize=16, pad=20)
plt.xlabel("客户数量", fontsize=12)
plt.ylabel("客户分群", fontsize=12)

#添加数值标签
for i,v in enumerate(df["customer_count"]):
    plt.text(v + 10, i, str(v), va="center", fontsize=11)

plt.tight_layout()
plt.savefig("D:/study/Python/data_analyst/data_clean/rfm_customer_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

### 各客群平均消费金额条形图

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=df, x="avg_monetary", y="Segment", palette="Greens_d",hue="Segment", legend=False)
plt.title('各客群平均消费金额 (Monetary)', fontsize=16, pad=20)
plt.xlabel('平均消费金额', fontsize=12)
plt.ylabel('客户群', fontsize=12)

#添加数值标签
for i, row in df.iterrows():
    plt.text(row['avg_monetary'], i, f"{int(row['avg_monetary'])}", 
             va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.savefig('D:/study/Python/data_analyst/data_clean/rfm_avg_monetary.png', dpi=300, bbox_inches='tight')
plt.show()
```
