# 这是一个电商分析的小项目

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#设置中文字体（防止乱码）
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

### Top 5 国家销量柱状图

```python
country = pd.read_csv("D:/study/Python/data_analyst/data_clean/country_sales.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=country, x="revenue", y="Country", hue="Country", palette="Blues_d", legend=False)
plt.title("Top 5 Countries by Revenue", fontsize=16)
plt.xlabel("Revenue", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.tight_layout()
plt.savefig("D:/study/Python/data_analyst/data_clean/country_revenue.png",dpi=300, bbox_inches="tight")
plt.show()
```

### 月销售趋势折线图

```python
monthly = pd.read_csv("D:/study/Python/data_analyst/data_clean/monthly_trend.csv")
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly, x="year_monthly", y="monthly_revenue", marker="o", linewidth=2.5, color="blue")
plt.title("Monthly Revenue Trend (2010)", fontsize=16)
plt.xlabel("Year-Month", fontsize=12)
plt.ylabel("Revenue",fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("D:/study/Python/data_analyst/data_clean/monthly_trend.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Top10 商品条形图

```python
products = pd.read_csv("D:/study/Python/data_analyst/data_clean/top_products.csv")
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=products.head(10), x="total_revenue", y="Description", hue="Description", palette="viridis", legend=False)
plt.title("Top 10 Best Selling Products by Revenue", fontsize=16)
plt.xlabel("Revenue", fontsize=12)
plt.ylabel("Product Description", fontsize=12)
plt.tight_layout()
plt.savefig("D:/study/Python/data_analyst/data_clean/top_products.png", dpi=300, bbox_inches="tight")
plt.show()
```
